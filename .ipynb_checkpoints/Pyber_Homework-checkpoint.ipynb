# Declarations

```python
%matplotlib inline
#%matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Read CSV Files and Combine into Single dataset

```python
# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
```

```python
# Read the City and Ride Data
City_Df = pd.read_csv(city_data_to_load, encoding = 'utf-8')
Ride_Df = pd.read_csv(ride_data_to_load, encoding ='utf-8')
```

```python
# Combine the data into a single dataset
City_Ride= Ride_Df.merge(City_Df, how = "left", on = "city")
# Display the data table for preview
City_Ride.head()
```

# DataFrame for Type + Urban


```python
City_Ride_Urban = City_Ride.loc[(City_Ride["type"]=="Urban")]
Cities_Urban = City_Ride_Urban["city"].unique()
City_Ride_Suburban = City_Ride.loc[(City_Ride["type"]=="Suburban")]
City_Ride_Rural = City_Ride.loc[(City_Ride["type"]=="Rural")]
Average_Fare_Urban = []
Total_Rides_Urban=[]
Total_Drivers_Urban=[]

for city in Cities_Urban:
    City_Urban = City_Ride_Urban.loc[(City_Ride_Urban["city"]== city),:]
    Average_Fare_Urban.append(City_Urban["fare"].mean())
    Total_Rides_Urban.append(len(City_Urban["ride_id"]))
    Total_Drivers_Urban.append(City_Urban["driver_count"].mean())
Urban = pd.DataFrame({"City": Cities_Urban,"Average Fare $ per City": Average_Fare_Urban,\
                      "Total Number of Rides":Total_Rides_Urban, "Drivers": Total_Drivers_Urban})
Urban.head()
```

# DataFrame for Type = Suburban

```python
City_Ride_Suburban = City_Ride.loc[(City_Ride["type"]=="Suburban")]
Cities_Suburban = City_Ride_Suburban["city"].unique()
Average_Fare_Suburban = []
Total_Rides_Suburban=[]
Total_Drivers_Suburban=[]

for city in Cities_Suburban:
    City_Suburban = City_Ride_Suburban.loc[(City_Ride_Suburban["city"]== city),:]
    Average_Fare_Suburban.append(City_Suburban["fare"].mean())
    Total_Rides_Suburban.append(len(City_Suburban["ride_id"]))
    Total_Drivers_Suburban.append(City_Suburban["driver_count"].mean())
Suburban = pd.DataFrame({"City": Cities_Suburban,"Average Fare $ per City": Average_Fare_Suburban,\
                      "Total Number of Rides":Total_Rides_Suburban, "Drivers": Total_Drivers_Suburban})
Suburban.head()
    
    
```

# DataFrame for Type = Rural

```python
City_Ride_Rural = City_Ride.loc[(City_Ride["type"]=="Rural")]
Cities_Rural = City_Ride_Rural["city"].unique()
Average_Fare_Rural = []
Total_Rides_Rural=[]
Total_Drivers_Rural=[]

for city in Cities_Rural:
    City_Rural = City_Ride_Rural.loc[(City_Ride_Rural["city"]== city),:]
    Average_Fare_Rural.append(City_Rural["fare"].mean())
    Total_Rides_Rural.append(len(City_Rural["ride_id"]))
    Total_Drivers_Rural.append(City_Rural["driver_count"].mean())
Rural = pd.DataFrame({"City": Cities_Rural,"Average Fare $ per City": Average_Fare_Rural,\
                      "Total Number of Rides":Total_Rides_Rural, "Drivers": Total_Drivers_Rural})
Rural.head()
```

# Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
x = Total_Rides_Urban
y= Average_Fare_Urban
z= np.array(Total_Drivers_Urban)
scale = z*10
```

```python
x_suburban = Total_Rides_Suburban
y_suburban= Average_Fare_Suburban
z_suburban= np.array(Total_Drivers_Suburban)
scale_sub = z_suburban*10
```

```python
x_rural = Total_Rides_Rural
y_rural= Average_Fare_Rural
z_rural= np.array(Total_Drivers_Rural)
scale_rural = z_rural*10
```

```python
# Build the scatter plots for each city types
#fig = plt.figure()
#ax1 = fig.add_subplot(111)
labels =["Urban","Suburban","Rural"]
Urban= plt.scatter(x, y, s=scale, alpha=0.85, marker="o", facecolors="coral", edgecolors="black",linewidth = 1)
Suburban= plt.scatter(x_suburban, y_suburban, s=scale_sub, alpha=0.85, marker="o", facecolors="skyblue", edgecolors="black",linewidth = 1)
Rural= plt.scatter(x_rural, y_rural, s=scale_rural, alpha=0.85, marker="o", facecolors="gold", edgecolors="black",linewidth=1)
# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)",fontsize=11)
plt.ylabel("Average Fare ($)",fontsize=11)
plt.ylim(19, 44.5)
plt.xlim(.1,41)
plt.grid()

# Create a legend
lgnd =plt.legend([Urban, Suburban,Rural],["Urban", "Suburban","Rural"],loc="upper right",
           scatteryoffsets=[0.5,0.5,0.5], markerscale=0.5,numpoints=1, title ="City Types", fontsize = 8) 
lgnd.legendHandles[0]._sizes = [40]
lgnd.legendHandles[1]._sizes = [40]
lgnd.legendHandles[2]._sizes = [40]
# Incorporate a text label regarding circle size
note= "Note:\nCircle size correlates with driver count per city"
plt.text(43, 35, note,fontsize=11)
# Save Figure
plt.savefig("../Images/Pyber_Ride_Sharing_Data.png")

# Show plot
plt.show()
plt.tight_layout()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
Total_Rides=(len(City_Ride_Urban["ride_id"])+len(City_Ride_Suburban["ride_id"])+len(City_Ride_Rural["ride_id"]))
Urban_Rides =len(City_Ride_Urban["ride_id"])*100/Total_Rides
Suburban_Rides = len(City_Ride_Suburban["ride_id"])*100/Total_Rides
Rural_Rides=len(City_Ride_Rural["ride_id"])*100/Total_Rides
Urban_Rides + Suburban_Rides + Rural_Rides
# Build Pie Chart

# The values of each section of the pie chart
sizes = [Urban_Rides,Suburban_Rides,Rural_Rides]
# The colors of each section of the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140,counterclock=False,pctdistance =0.6)
plt.title("% of Total Rides by City Type")
plt.axis("normal")
# Save Figure
plt.savefig("../Images/Total_Rides_By_City_Type.png")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
City_Drivers=[]
sum=0
labels=["Urban","Suburban","Rural"]
City_Ride_Drivers=City_Ride.loc[:,["city","driver_count","type"]].drop_duplicates()
for label in labels:
    Type_drivers = City_Ride_Drivers.loc[City_Ride_Drivers["type"]==label]
    total=Type_drivers["driver_count"].sum()
    sum=sum+total
    City_Drivers.append(total)
Urban_Drivers=City_Drivers[0]*100/sum
Suburban_Drivers=City_Drivers[1]*100/sum
Rural_Drivers=City_Drivers[2]*100/sum
sizes = [Urban_Drivers,Suburban_Drivers,Rural_Drivers]
# The colors of each section of the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=160, counterclock =False,pctdistance =0.7)
plt.title("% of Total Drivers by City Type")
plt.axis("normal")
# Save Figure
plt.savefig("../Images/Total_Drivers_By_City_Type.png")
```

```python
# Show Figure
plt.show()
```
